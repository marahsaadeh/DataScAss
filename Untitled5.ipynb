```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df=pd.read_csv('yellow_tripdata_2019-02 (1).csv')
```

```python
def trip_duration(df):
    df['The_trip_duration_min'] =  pd.to_datetime(df['tpep_dropoff_datetime']) - pd.to_datetime(df['tpep_pickup_datetime'])    
    return df
df =trip_duration(df)
```

```python
df
```

```python
def add_hour_day(df):
    df['hour_of_day'] = pd.to_datetime(df['tpep_pickup_datetime']).apply(lambda x: x.strftime("%H"))
    df['day_of_week'] = pd.to_datetime(df['tpep_pickup_datetime']).apply(lambda x: x.strftime("%A"))
    return df
df = add_hour_day(df)
```

```python
df
```

```python
time_trip_duration=df['The_trip_duration_min'].dt.components
df["time_trip_duration"]=time_trip_duration["minutes"]
```

```python
def compute_predictions(df):
    predictions = df.groupby(['PULocationID', 'DOLocationID', 'day_of_week', 'hour_of_day'])['time_trip_duration']
    mean = predictions.mean()
    standard_deviation = predictions.std()
    t_score_value= stats.t.ppf(0.975, len(df) - 1)
    pred = pd.DataFrame({'mean_trip_duration': mean, 'margin_of_error': t_score_value * standard_deviation / np.sqrt(len(df))})
    return pred
```

```python
z_value = stats.norm.ppf(0.9505)
z_value
```
